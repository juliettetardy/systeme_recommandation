# Visualisation des donn√©es

```python
import pandas as pd
import matplotlib.pyplot as plt
import json

# Charger le fichier JSON 
with open('etiquetage.json', 'r') as file:
    data = json.load(file)

# Convertir les donn√©es JSON en DataFrame
df = pd.DataFrame(data)

# 1. Analyser les fr√©quences des diff√©rentes caract√©ristiques
grouped_orientation = df.groupby('orientation').size()
grouped_qualit√© = df.groupby('qualite').size()
grouped_animal = df.groupby('animal').size()
grouped_nature = df.groupby('nature').size()
grouped_luminosit√© = df.groupby('luminosite').size()

# 2. Visualisation des distributions avec matplotlib

# Orientation
plt.figure(figsize=(5, 3))
grouped_orientation.plot(kind='bar', color="lightgreen")
plt.title("Distribution des images par orientation")
plt.xlabel("Orientation")
plt.ylabel("Nombre d'images")
plt.xticks(rotation=0)
plt.show()

# Qualit√©
plt.figure(figsize=((5, 3)))
grouped_qualit√©.plot(kind='bar', color="lightblue")
plt.title("Distribution des images par qualit√©")
plt.xlabel("Qualite")
plt.ylabel("Nombre d'images")
plt.xticks(rotation=0)
plt.show()

# Animal
plt.figure(figsize=(5, 3))
grouped_animal.plot(kind='bar', color="lightcoral")
plt.title("Distribution des images par animal")
plt.xlabel("Animal")
plt.ylabel("Nombre d'images")
plt.xticks(rotation=0)
plt.show()

# Nature
plt.figure(figsize=(5, 3))
grouped_nature.plot(kind='bar', color="lightyellow")
plt.title("Distribution des images par nature")
plt.xlabel("Nature")
plt.ylabel("Nombre d'images")
plt.xticks(rotation=0)
plt.show()

# Luminosit√©
plt.figure(figsize=(5, 3))
grouped_luminosit√©.plot(kind='bar', color="lightskyblue")
plt.title("Distribution des images par luminosit√©")
plt.xlabel("Luminosite")
plt.ylabel("Nombre d'images")
plt.xticks(rotation=0)
plt.show()
```
